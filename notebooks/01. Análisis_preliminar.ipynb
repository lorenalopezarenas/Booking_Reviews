Antes de comenzar el análisis exploratorio de datos (EDA), es importante realizar una inspección general de los datasets. ## `booking_raw`

```python
# Llamamos a la función eda_preliminar guardada en el archivo de sorporte
eda_preliminar(booking_raw)
```

### **Informe de EDA preliminar – Dataset `booking_raw`**

**1. Información general del Dataset**

- El Dataset contiene 119390 registros y +30 variables

- Incluye columnas numéricas, categóricas y temporales.

- Se identifican valores nulos en `agent`, `company`, `country` y `children`.

- No presenta registros duplicados → dataset consistente.


**2. Hallazgos por tipo de variable**

Variables numéricas:

- `adr`: alta variabilidad (hasta 5400), indicando posibles outliers.
- `lead_time`: media de 104 días, reflejando reservas con bastante anticipación.
- `is_canceled`: media del 37% → alta tasa de cancelación.

Variables categóricas:

- `adults`, `children`, `babies`: predominan reservas de 1–2 adultos sin niños.
- `previous_cancellations`: valores bajos → pocos clientes reincidentes cancelando.
- `reserved_room_type` y `assigned_room_type` → las mantenemos para analizar posibles discrepancias entre la habitación reservada y la asignada, que pueden correlacionar con reviews negativas.
- Predominan clientes “transient” y reservas online.


**3. Transformaciones clave**

Estandarización:
- Variables categóricas → minúsculas + snake_case.
- `phone-number` → `phone_number`.

Eliminación:

- `arrival_date_week_number` y `arrival_date_day_of_month`

    Solo necesitamos el mes y año de llegada para poder unir este dataset con el de reviews. Por lo tanto, estas columnas se pueden eliminar.

- `agent` y `company`

    Representan los ID de las agencias y compañías. No aportan información relevante para nuestro análisis, por lo que se eliminarán.

- `days_in_waiting_list`

    La mayoría de reservas no pasan por lista de espera; además, esta variable describe un proceso interno del hotel, no tanto al cliente. Por lo que la variable aporta poca información.

- `required_car_parking_spaces` y `total_of_special_requests`

    El valor promedio es muy bajo, la desviación estándar es baja y los percentiles muestran que la mayoría de los huéspedes no requiere estas opciones. Estas columnas aportan poca información útil y se eliminan.

- `name`, `email`, `phone-number`, `credit_card` 

    Nos dan información sensible del huésped que no es que no aporta valor al análisis. 


Tipos de datos:
- `is_canceled`, `is_repeated_guest` → bool.
- `children` → int.
- `reservation_status_date` → datetime.      

Feature Engineering:

- `total_nights` → Número de noches totales. 
- `total_guests` → Total de personas por reserva. 


💡 **Conclusión:**

El dataset está bien estructurado y, tras limpieza y transformación, queda listo para análisis exploratorio avanzado.

## `reviews_raw`

```python
# Llamamos a la función eda_preliminar guardada en el archivo de sorporte
eda_preliminar(reviews_raw)
```

### **Informe de EDA preliminar – Dataset `reviews_raw`**

**1. Información general del Dataset**

- El dataset contiene 515738 registros y 17 variables.

- Presenta un bajo porcentaje de valores nulos en `lat` y `lng` (~0.63%).

- Existen algunos duplicados, pero la estructura general es consistente.

**2. Hallazgos por tipo de variable**

Variables numéricas:

- `Reviewer_Score` y `Average_Score`: media ≈ 8.4 → alto nivel de satisfacción general.
- `Total_Number_of_Reviews`: gran dispersión → diferentes niveles de popularidad de los hoteles.

Variables de texto y categóricas:

- `Positive_Review` y `Negative_Review`: alta variabilidad en contenido.
- `Reviewer_Nationality`: predominan usuarios del Reino Unido, posible sesgo geográfico.

**3. Transformaciones clave**

Estandarización:

- Nombres de las columnas → minúsculas.
- Normalización de `reviewer_nationality`: snake_case para homogeneizar valores (por ejemplo, *united_kingdom*).

Eliminación:

- `review_total_negative_word_counts` y `review_total_positive_word_counts`: no aportan valor para nuestro análisis.
- `days_since_review`: variable dependiente del momento de extracción, sin valor analítico.
- `lat` y `lng`: redundancia, ya tenemos la localización.

Tipos de datos:

- `review_date` → datetime.

Feature engineering:

- `city` y `country` → a partir de la dirección del hotel.
- Simplificación de `positive_review` y `negative_review` a palabras clave.
- `trip_type`, `group_type`, `room_type`, `nights` y `device` → a partir de la columna `tags`. 

💡 Conclusión:

El dataset es rico en información y adecuado para análisis de sentimiento. Tras limpieza, estandarización y creación de nuevas variables, queda optimizado para análisis avanzado e integración con bookings.
