## `booking_raw`

```python
# Llamamos a la función eda_preliminar guardada en el archivo de sorporte
eda_preliminar(booking_raw)
```

### **Informe de EDA preliminar – Dataset `booking_raw`**

**1. Información general del Dataset**

- El Dataset contiene 119390 registros y +30 variables

- Incluye columnas numéricas, categóricas y temporales.

- Se identifican valores nulos en `agent`, `company`, `country` y `children`.

- No presenta registros duplicados → dataset consistente.

**2. Hallazgos por tipo de variable**

Variables numéricas:

- `adr`: alta variabilidad (hasta 5400), indicando posibles outliers.
- `lead_time`: media de 104 días, reflejando reservas con bastante anticipación.
- `is_canceled`: media del 37% → alta tasa de cancelación.

Variables categóricas:

- `adults`, `children`, `babies`: predominan reservas de 1–2 adultos sin niños.
- `previous_cancellations`: valores bajos → pocos clientes reincidentes cancelando.
- Predominan clientes “transient” y reservas online.

**3. Transformaciones clave**

Estandarización:
- `hotel`, `customer_type`, `reservation_status` → minúsculas + snake_case.
- `phone-number` → `phone_number`.

Tipos de datos:
- `is_canceled`, `is_repeated_guest` → bool.
- `reservation_status_date` → datetime.

Eliminación:
- `name`, `email`, `phone_number`, `credit_card` → información del huésped que no es relevante para nuestro análisis.
- `arrival_date_week_number` → ya disponemos de los días, meses y años de llegada. 
- `assigned_room_type` → Podría mantenerse para analizar discrepancias entre la habitación reservada y la asignada, que pueden correlacionar con reviews negativas.

💡 **Conclusión:**

El dataset está bien estructurado y, tras limpieza y transformación, queda listo para análisis exploratorio avanzado.

### **Informe de EDA preliminar – Dataset `reviews_raw`**

**1. Información general del Dataset**

- El dataset contiene 515738 registros y 17 variables.

- Presenta pocos valores nulos en lat y lng (~0.63%).

- Existen algunos duplicados, pero la estructura general es consistente.

**2. Hallazgos por tipo de variable**

Variables numéricas:

- Reviewer_Score y Average_Score: media ≈ 8.4 → alta satisfacción general.
- Total_Number_of_Reviews: gran dispersión → diferentes niveles de popularidad.

Variables de texto y categóricas:

- Positive_Review y Negative_Review: alta variabilidad en contenido.
- Reviewer_Nationality: predominan usuarios del Reino Unido.

**3. Transformaciones clave**

Estandarización:

- Nombres de las columnas → minúsculas + snake_case.
- reviewer_nationality: “United Kingdom” → “UK”.
- tags: limpiar caracteres innecesarios.

Tipos de datos:

- review_date → datetime.
- days_since_review → eliminar la palabra “day” y convertir a int.

Feature engineering:

- hotel_address → crear city y country.
- Reducir cada review en positive_review y negative_review a una “palabra clave”

Eliminación:

. review_total_negative_word_counts y review_total_positive_word_counts.
- lat y lng: redundancia, ya tenemos la localización.

💡 Conclusión:

El dataset es rico en información y adecuado para análisis de sentimiento. Tras limpieza, estandarización y creación de nuevas variables, queda optimizado para análisis avanzado e integración con bookings.
